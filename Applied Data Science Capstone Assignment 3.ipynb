### The read_html method of pandas gets all the tables in the html page. The given wikipedia page contains 3 tables which are returned in a list. The table that is required for this assignment is at first index position in the list

```python
df = a[0]
```

```python
df
```

### Changing all the Not assigned values in the Borough column to nan values so that we can use isna() to check for nan values and eliminate cells that have these values

```python
df['Borough'] = df['Borough'].apply(lambda x: np.nan if x=='Not assigned' else x)
```

```python
df1= df['Borough'].isna()
```

```python
df = df[~df1]
```

### Replacing the neighborhoods who are Not assigned with their Borough Names

```python
df['Neighborhood']=df['Neighborhood'].replace('Not assigned', df['Borough'])
```
